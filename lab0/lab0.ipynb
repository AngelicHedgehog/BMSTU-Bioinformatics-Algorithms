```python
import string
from heapq import nlargest
```

```python
data = {}

with open('uniprot_sprot.fasta', 'r') as f:
    last_line = f.readline()
    i = 0
    while last_line != '':
        head = last_line[:last_line.find('=') - 3]

        last_line = f.readline()
        gen = ''
        while last_line != '' and last_line[0] != '>':
            gen += last_line[:-1]
            last_line = f.readline()
        
        alphabet = dict(zip(string.ascii_uppercase, [0] * 26))
        for a in gen:
            alphabet[a] += 1
        
        data[head] = alphabet

        if i % 50000 == 0:
            print(i)
        i += 1
```

```python
def find_nearest(DNK):
    alphabet = dict(zip(string.ascii_uppercase, [0] * 26))
    for a in DNK:
        alphabet[a] += 1

    koefs = {}
    for head in data:
        koef = 0
        for alpha in alphabet:
            a = alphabet[alpha]
            b = data[head][alpha]
            c = min(a, b)
            koef += c / (a + b - c) if not (a == b == 0) else 1
        koef /= len(alphabet)

        koefs[head] = koef
    
    return nlargest(100, koefs, key=lambda x: koefs[x])
```
